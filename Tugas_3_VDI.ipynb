Tujuan: Membuat grafik perubahan harga dari tahun 2019 hingga 2024 serta perubahan volumenya

User: Data tersebut ditujukan untuk orang yang ingin mengetahui perkembangan saham Apple (AAPL)

Fungsi:
*   Date = Tanggal tercatat saham
*   Open = Harga awal dibuka di hari tersebut
*   High = Harga tertinggi di hari tersebut
*   Low = Harga terendah di hari tersebut
*   Close = Harga akhir ditutup di hari tersebut
*   Volume = Jumlah saham yang beredar




```python
import pandas as pd
import plotly.graph_objects as go
from datetime import datetime
```

```python
data = pd.read_csv("/content/aapl_us_d.csv")
data
```

```python
data.isnull().sum()
```

```python
Viz = go.Figure(data=[go.Candlestick(x=data['Date'],
                open=data['Open'],
                high=data['High'],
                low=data['Low'],
                close=data['Close'])])

Viz.show()
```

```python
VolViz = go.Figure([go.Scatter(x=data['Date'], y=data['Volume'])])
VolViz.show()
```
